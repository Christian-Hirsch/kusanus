Zunaechst betrachten wir fuer ein Spielzeugbeispiel, wie sich convolutions auf Matrixebene auswirken. Dazu benoetigen wir eine  (3,3) Matrix. Diese kann entweder frei gewaehlt werden, oder mit dem Befehl `np.random.randint(???)` erzeugt werden. Erinnere dich an die Tools aus dem Jupyter-Notebook cheat sheet.

```python
import numpy as np
from scipy.signal import correlate

np.random.seed(42)

matrix = np.random.randint()
print(matrix)
img_filter = [[1, -1], [1,2]]

correlate(matrix, img_filter, mode = 'same')
```
